```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.simplefilter('ignore') # ignore warnings
data = pd.read_csv('../data/raw/countries.csv')
```

```python
data.head()
```

```python
# scatter life expectancy and gdp per cap
scatter = plt.scatter(
    np.log10(data.gdpPerCapita)
    ,data.lifeExpectancy
    ,10
    ,c=data.year # coloring by year
)
plt.legend(*scatter.legend_elements())
plt.title('gdp per cap and life expectancy by country')
plt.xlabel('log gdp per cap')
plt.ylabel('life expectancy')
plt.show()
```

```python
# scatter life expectancy and gdp per cap per year
for year in set(data.year):
    iter_data = data[data.year==year]
    plt.scatter(
        np.log10(iter_data.gdpPerCapita)
        ,iter_data.lifeExpectancy
        ,5
    )
    plt.title(f'{year} gdp per cap and life expectancy by country')
    plt.xlabel('log gdp per cap')
    plt.ylabel('life expectancy')
    plt.show()
```

```python
# add gdp
data.gdp = data.population * data.gdpPerCapita
```

```python
# scatter life expectancy and gdp non log for 2007
data_2007 = data[data.year == 2007]
plt.scatter(data_2007.gdp,data_2007.lifeExpectancy,10)
plt.title(f'2007 gdp and life expectancy by country')
plt.xlabel('gdp')
plt.ylabel('life expectancy')
plt.show()
```

```python
# scatter life expectancy and gdp log for 2007
data_2007 = data[data.year == 2007]
plt.scatter(np.log10(data_2007.gdp),data_2007.lifeExpectancy,10)
plt.title('2007 gdp and life expectancy by country')
plt.xlabel('log gdp')
plt.ylabel('life expectancy')
plt.show()
```
